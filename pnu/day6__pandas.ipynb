 ## 5. Inserting and Deleting Rows and Columns

```python
df = pd.read_csv('https://raw.githubusercontent.com/agconti/kaggle-titanic/master/data/train.csv')
df.head()
```

```python
first = lambda x: x[0]
def getFirstName(df): # firstname만 따로 뽑고싶다.
  return df.Name.str.replace(',','').str.split().apply(first) # str 감각 apply -> map

df2 = df.copy() # 원본 데이터 건드리지 말기
df2['FirstName'] = getFirstName(df) # 컬럼 생성(붙이기) -> FirstName
df2.head()
```

```python
# 원하는 자리에 붙이기.
df3 = df.copy()
df3.insert(3, 'FirstName', getFirstName(df))
df3.head()
```

```python
# 지우기 원본은 안바뀜!
droped_df = df3.drop(['FirstName'], axis=1)
droped_df.head() 
```

##  10. Missing Data

```python
import pandas as pd

df = pd.read_csv('https://raw.githubusercontent.com/agconti/kaggle-titanic/master/data/train.csv')
df.info()
```

```python
df.isna() # 활용해서 불리언 인덱싱을 할 수 있음.
```

```python
df[df.isna().any(axis=1)] # 응용한거
```

```python
# ffill: propagate last valid observation forward to next valid.
# backfill / bfill: use next valid observation to fill gap.

df_clean = df.fillna(method='ffill')
df_clean = df_clean.fillna(method='backfill')
NA_MASK=df_clean[df_clean.isna().any(axis=1)]
df_clean[NA_MASK]
# isna 물어보는거 fillna 채우는것
```

```python
df_clean = df['Age'].fillna(df['Age'].mean())
df_clean.info()
```

```python
df['Embarked'].isna()
```

```python
df_clean.mean()
```

```python
df_clean2 = df.fillna(df.mean())
df_clean2.mean()
```
